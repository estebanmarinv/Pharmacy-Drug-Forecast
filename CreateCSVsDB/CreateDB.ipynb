```python
#DATABASE CONNECTION
import pandas as pd
import os
from sqlalchemy import create_engine, text

#maximum number of rows to display
pd.options.display.max_rows = 20

engine=create_engine('postgresql://postgres:Ds4_MX_2020@54.184.252.64/DalindeDB', max_overflow=30)

def runQuery(sql):
    result = engine.connect().execution_options(isolation_level="AUTOCOMMIT").execute((text(sql)))
    return pd.DataFrame(result.fetchall(), columns=result.keys())

def setup():
    #Load CSVs FIles
    PatentDrugs_file = os.path.abspath('/users/shared/PatentDrug.csv')
    Requests_file = os.path.abspath('/users/shared/Request.csv')
    RequestsMedCode_file = os.path.abspath('/users/shared/RequestMedCode.csv')
    PurchaseOrders_file = os.path.abspath('/users/shared/PurchaseOrder.csv')
    POReqMedCode_file = os.path.abspath('/users/shared/POReqMedCode.csv')
    PatientConsumption_file = os.path.abspath('/users/shared/PatientConsumption.csv')
    PatConsMedCode_file = os.path.abspath('/users/shared/PatConsMedCode.csv')
    
    #Create Tables in Database
    return runQuery("""
    CREATE TABLE "PatentDrug" ("MedCode" BIGINT, "MedDescription" CHARACTER VARYING(100), "MedPharmacon" CHARACTER VARYING(50), "MedFamily" CHARACTER VARYING(80));
    CREATE TABLE "Request" ("RequestNumber" INT, "RequestDate" date);
    CREATE TABLE "RequestMedCode" ("RequestNumber" INT, "MedCode" BIGINT, "AmountRequested" INT);
    CREATE TABLE "PurchaseOrder" ("PurchaseNumber" INT, "OrderDate" date, "RequiredDeliveryDate" date, "SupplierName" CHARACTER VARYING(80));
    CREATE TABLE "POReqMedCode" ("PurchaseNumber" INT, "RequestNumber" INT, "MedCode" BIGINT, "AmountPurchased" INT);
    CREATE TABLE "PatientConsumption" ("ConsumptionNumber" INT, "ConsumptionDate" DATE, "Department" CHARACTER VARYING(20));
    CREATE TABLE "PatConsMedCode" ("ConsumptionNumber" INT, "MedCode" BIGINT, "AmountConsumed" INT);
    COPY "PatentDrug" FROM '""" + PatentDrugs_file + """' WITH (format csv, header true, delimiter ',');
    COPY "Request" FROM '""" + Requests_file + """' WITH (format csv, header true, delimiter ',');
    COPY "RequestMedCode" FROM '""" + RequestsMedCode_file + """' WITH (format csv, header true, delimiter ',');
    COPY "PurchaseOrder" FROM '""" + PurchaseOrders_file + """' WITH (format csv, header true, delimiter ',');
    COPY "POReqMedCode" FROM '""" + POReqMedCode_file + """' WITH (format csv, header true, delimiter ',');
    COPY "PatientConsumption" FROM '""" + PatientConsumption_file + """' WITH (format csv, header true, delimiter ',');
    COPY "PatConsMedCode" FROM '""" + PatConsMedCode_file + """' WITH (format csv, header true, delimiter ',');
    
    SELECT * FROM "PurchaseOrder" LIMIT(5);""")

def cleanup():
    #Erase Tables of Database
    runQuery("""
    DROP TABLE "PatentDrug" CASDCADE;
    DROP TABLE "Request" CASCADE;
    DROP TABLE "RequestMedCode" CASCADE;
    DROP TABLE "PurchaseOrder" CASCADE;
    DROP TABLE "POReqMedCode" CASCADE;
    DROP TABLE "PatientConsumption" CASCADE;
    DROP TABLE "PatConsMedCode" CASCADE;    
    SELECT 0 WHERE FALSE; -- prevents SQLAlchemy from throwing an error;
    """)
```
