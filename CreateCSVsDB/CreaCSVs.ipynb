```python
import pandas as pd

#Read CSVs files provided for Dalinde
req_compras = pd.read_csv('PurchasedOrders_2018-2019.csv')
pat_consumption = pd.read_csv('PatientConsumption 2018-2019.csv')
```

```python
#CREATE "Requests" file
Request = pd.DataFrame()
Request = req_compras[['RequestNumber','RequestDate']].copy()
Request = Request.drop_duplicates()
Request.set_index('RequestNumber')
Request = Request.sort_values(by='RequestNumber', ascending=True)
Request.drop(Request[Request.RequestNumber.isnull()].index, inplace = True)
```

```python
Request.head()
```

```python
Request.to_csv('Request.csv', index=False)
```

```python
#CREATE "RequestsMedCode" file
ReqMedCode = pd.DataFrame()
ReqMedCode = req_compras[['RequestNumber','MedCode','AmountRequested']].copy()
ReqMedCode = ReqMedCode.drop_duplicates()
ReqMedCode.set_index(['RequestNumber', 'MedCode'])
ReqMedCode = ReqMedCode.sort_values(by=['RequestNumber','MedCode'], ascending=True)
ReqMedCode['AmountRequested']=ReqMedCode['AmountRequested'].apply(lambda x: round(x))
```

```python
ReqMedCode.head()
```

```python
ReqMedCode.to_csv('RequestMedCode.csv', index=False)
```

```python
#CREATE "PurchasesOrder" file
PO = pd.DataFrame()
PO = req_compras[['PurchaseNumber','OrderDate','RequiredDeliveryDate','SupplierName']].copy()
PO = PO.drop_duplicates()
PO.set_index(['PurchaseNumber'])
PO = PO.sort_values(by=['PurchaseNumber'], ascending=True)
PO.drop(PO[PO.PurchaseNumber.isnull()].index, inplace = True)
PO['PurchaseNumber']=PO['PurchaseNumber'].apply(lambda x: round(x))
```

```python
PO.head()
```

```python
PO.tail()
```

```python
PO.to_csv('PurchaseOrder.csv', index=False)
```

```python
#CREATE "POReqMedCode" file
POReqMedCode = pd.DataFrame()
POReqMedCode = req_compras[['PurchaseNumber','RequestNumber','MedCode','AmountPurchased']].copy()
POReqMedCode = POReqMedCode.drop_duplicates()
POReqMedCode.set_index(['PurchaseNumber','RequestNumber', 'MedCode'])
POReqMedCode = POReqMedCode.sort_values(by=['PurchaseNumber','RequestNumber','MedCode'], ascending=True)
POReqMedCode['AmountPurchased']=POReqMedCode['AmountPurchased'].apply(lambda x: round(x))
POReqMedCode.drop(POReqMedCode[POReqMedCode.RequestNumber.isnull()].index, inplace = True)
POReqMedCode.drop(POReqMedCode[POReqMedCode.PurchaseNumber.isnull()].index, inplace = True)
POReqMedCode['PurchaseNumber']=POReqMedCode['PurchaseNumber'].apply(lambda x: round(x))
```

```python
POReqMedCode.to_csv('POReqMedCode.csv', index=False)
```

```python
#CREATE "PatientConsumption" file
PatientConsumption = pd.DataFrame()
PatientConsumption = pat_consumption[['ConsumptionNumber','ConsumptionDate','Department']].copy()
PatientConsumption = PatientConsumption.sort_values(by=['ConsumptionNumber'], ascending=True)
PatientConsumption['ConsumptionNumber']=PatientConsumption['ConsumptionNumber'].apply(lambda x: round(x))
PatientConsumption = PatientConsumption.drop_duplicates(subset="ConsumptionNumber", keep='first', inplace=False)
```

```python
PatientConsumption
```

```python
PatientConsumption.to_csv('PatientConsumption.csv', index=False)
```

```python
#CREATE "PatientConsumption" file
PatConsMedCode = pd.DataFrame()
PatConsMedCode = pat_consumption[['ConsumptionNumber','MedCode', 'AmmountConsumed']].copy()
PatConsMedCode.set_index(['ConsumptionNumber', 'MedCode'])
PatConsMedCode = PatConsMedCode.drop_duplicates()
PatConsMedCode = PatConsMedCode.sort_values(by=['ConsumptionNumber', 'MedCode'], ascending=True)
PatConsMedCode['ConsumptionNumber']=PatConsMedCode['ConsumptionNumber'].apply(lambda x: round(x))
PatConsMedCode.drop(PatConsMedCode[PatConsMedCode.ConsumptionNumber.isnull()].index, inplace = True)
PatConsMedCode['MedCode']=PatConsMedCode['MedCode'].apply(lambda x: round(x))
PatConsMedCode.drop(PatConsMedCode[PatConsMedCode.MedCode.isnull()].index, inplace = True)
PatConsMedCode['AmmountConsumed']=PatConsMedCode['AmmountConsumed'].apply(lambda x: round(x))
PatConsMedCode.drop(PatConsMedCode[PatConsMedCode.AmmountConsumed.isnull()].index, inplace = True)
```

```python
PatConsMedCode.to_csv('PatConsMedCode.csv', index=False)
```
