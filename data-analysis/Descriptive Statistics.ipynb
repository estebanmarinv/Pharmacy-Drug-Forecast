```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
import os
import seaborn as sns
```

```python
#Read file 
consumption = pd.read_csv("Informe de consumo 2018-2019.csv")
orders= pd.read_csv("Requisiciones vs ordenes de compra 2018-2019.csv")
orders["FECHA"]=pd.to_datetime(orders["FECHA"], format="%YYY-")
```

```python
desabastos=orders[orders["CANTIDAD PEDIDA"]==0]


plt.figure(figsize=(15, 5))

plt.hist(desabastos.FECHA, bins = 150, facecolor="m")

plt.title("Frequency of ratings for Videos over time")
plt.xlabel("Rating")
plt.ylabel("Frecuency of Rating")
plt.grid(True)
plt.show()
```
